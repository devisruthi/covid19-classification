```python
import pandas as pd
import cv2
from pathlib import Path
import os
```

```python
datapath = Path("../labelled-csv/train.csv")
df_data = pd.read_csv(datapath)
df_data.head(2)
```

```python
train_path = Path("../../covidx-cxr2-dataset/train")
```

```python


# Load all images
count = 0
for imagepath in train_path.glob("*.*"):
    if count >= 3:
        #break
        print(count)
    # Load the image from disk
    try:
        imagename = os.path.basename(imagepath)
        filedata_in_dataset = None
        labeldesc = None
        filedata_in_dataset = df_data[df_data['filepath'] == imagename].classification.values
        if len(filedata_in_dataset) == 1:
            print(count)
            labeldesc = filedata_in_dataset[0]
            img = cv2.imread(str(imagepath),0) 
            
            #blurred_image =  cv2.GaussianBlur(img,(5,5),0)
            # create a CLAHE object (Arguments are optional).
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
            cl1 = clahe.apply(img)
            
            basepath = Path("./clahe-train")
            saveto = Path(basepath)
            saveto = os.path.join(basepath, imagename)
            print(saveto)
            cv2.imwrite(saveto,cl1)
            count = count + 1
        else:
             continue
    except Exception as e: 
        print(e)
        #ignore if image not loaded or label not correct with printing name
        
```
